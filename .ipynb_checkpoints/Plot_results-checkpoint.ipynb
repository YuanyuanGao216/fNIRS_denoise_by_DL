```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.io as sio
from os import path
print('Packages imported successfully!')
```

```python
# load data

Real_HbO = sio.loadmat('Processed_data/Real_HbO.mat')['Real_HbO']
HbO_Spline = sio.loadmat('Processed_data/HbO_Spline.mat')['HbO_Spline']
HbO_Wavelet05 = sio.loadmat('Processed_data/HbO_Wavelet05.mat')['HbO_Wavelet05']
HbO_Wavelet40 = sio.loadmat('Processed_data/HbO_Wavelet40.mat')['HbO_Wavelet40']
HbO_Kalman = sio.loadmat('Processed_data/HbO_Kalman.mat')['HbO_Kalman']
HbO_PCA99 = sio.loadmat('Processed_data/HbO_PCA99.mat')['HbO_PCA99']
HbO_PCA50 = sio.loadmat('Processed_data/HbO_PCA50.mat')['HbO_PCA50']
HbO_Cbsi = sio.loadmat('Processed_data/HbO_Cbsi.mat')['HbO_Cbsi']

filepath = "Processed_data/Real_NN_8layers.mat"

if path.exists(filepath):
    Hb_NN = sio.loadmat(filepath)['Y_real']
else:
    Hb_NN = np.zeros(Real_HbO.shape)

HbO_NN = Hb_NN[:,0:Real_HbO.shape[1]];
HbR_NN = Hb_NN[:,Real_HbO.shape[1]:];

Real_HbR = sio.loadmat('Processed_data/Real_HbR.mat')['Real_HbR']
HbR_Spline = sio.loadmat('Processed_data/HbR_Spline.mat')['HbR_Spline']
HbR_Wavelet05 = sio.loadmat('Processed_data/HbR_Wavelet05.mat')['HbR_Wavelet05']
HbR_Wavelet40 = sio.loadmat('Processed_data/HbR_Wavelet40.mat')['HbR_Wavelet40']
HbR_Kalman = sio.loadmat('Processed_data/HbR_Kalman.mat')['HbR_Kalman']
HbR_PCA99 = sio.loadmat('Processed_data/HbR_PCA99.mat')['HbR_PCA99']
HbR_PCA50 = sio.loadmat('Processed_data/HbR_PCA50.mat')['HbR_PCA50']
HbR_Cbsi = sio.loadmat('Processed_data/HbR_Cbsi.mat')['HbR_Cbsi']

MA_list = sio.loadmat('Processed_data/MA_list.mat')['MA_list']
```

```python
#  figure 1: Example plot
# piece = input('Which piece you want to show? ')
piece = 2;
plt.figure(figsize=(100, 100), dpi= 80, facecolor='w', edgecolor='k')
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(Real_HbO[piece,:],label='No correction')
ax1.plot(HbO_Spline[piece,:])
ax1.plot(HbO_Wavelet05[piece,:])
ax1.plot(HbO_Wavelet40[piece,:])
ax1.plot(HbO_Kalman[piece,:])
ax1.plot(HbO_PCA99[piece,:])
ax1.plot(HbO_PCA50[piece,:])
ax1.plot(HbO_Cbsi[piece,:])
ax1.plot(HbO_NN[piece,:])
ax1.set_ylabel('$\Delta$ HbO ($\mu$Mol)')
ax1.legend()
plt.show()
```
